```python
import pandas as pd
csv_file_path = 'housing.csv'
column_names = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT","MEDV"]
df = pd.read_csv(csv_file_path, names=column_names)

df.head()
```

```python
df1 = df.rename(columns={"MEDV":"PRICE"})
df1.head()
```

```python
df1.corr()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.scatter(df1["RM"], df1["PRICE"], marker='o', color='blue', alpha=0.7, s=50, label='Room vs Price')

plt.xlabel('RM)')
plt.ylabel('Housing Price (PRICE)')
plt.title('Scatter Plot of Room vs Housing Price')

# Add grid
plt.grid(True)

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
import seaborn as sns
sns.regplot(x = df1["RM"], y = df1["PRICE"])
```

```python
X = df1.drop(["PRICE"], axis="columns")
y = df1["PRICE"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y, test_size=0.2)

```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,y_train)
```

```python
print(model.coef_)
```

```python
print(model.intercept_)
```

```python
pred = model.predict(X_test)
```

```python
plt.scatter(y_test,pred)
```

```python
residuals = y_test - pred #Residuals are errors
print(residuals)
```

```python
sns.displot(residuals, kind="kde")
```

```python
plt.scatter(pred,residuals)
```

```python
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print("Mean Absoulte Error =" , mean_absolute_error(y_test,pred))
print("Mean Squared Error =" , mean_squared_error(y_test,pred))
print("Root Mean Squared Error = " ,np.sqrt(mean_squared_error(y_test,pred)))
```

```python
from sklearn.metrics import r2_score
score = r2_score(y_test,pred)
print(score)
```

```python
#Adjust R2 Score
1 - (1-score)*(len(y_test)-1)/(len(y_test)- X_test.shape[1]-1)
```
